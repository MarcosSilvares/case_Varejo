```python
# Importando as Bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import cufflinks as cf
import plotly.graph_objects as go
import plotly.express as px
```

```python
# Trazendo as bases

vendas = pd.read_excel('/content/varejo.xlsx')
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
vendas.head()
```

```python
# Explorando mais o DataSet atravs da funo "info"

vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'] = vendas['estado'].fillna('MS')
```

```python
vendas.info()
```

```python
media_preo = vendas['Preo'].mean()

vendas['Preo'] = vendas['Preo'].fillna(media_preo)
```

```python
vendas.info()
```

```python
vendas[['Preo', 'Preo_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preo > Preo_com_frete')

vendas_correto = vendas.query('Preo < Preo_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
# Departamentos mais Vendidos

departamentos_vendas = vendas_correto.groupby('Nome_Departamento')['idcompra'].nunique().sort_values(ascending=False).reset_index()
departamentos_vendas.head(10)
```

```python
# Mdia de preo com frete por Nome de Departamento

agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')['Preo_com_frete'].mean().sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
# Quantidade de vendas por Ms

vendas_correto['mes'] = vendas_correto['Data'].dt.month # Criando coluna 'Ms'


vendas_mes = vendas_correto.groupby('mes')['idcompra'].nunique().sort_values(ascending=False).reset_index()
vendas_mes
```

```python
cliente.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
# Mdia de renda para cada tipo de canal de venda

agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].mean().sort_values(ascending=False).reset_index(), 2)
agg_idcanal_renda
```

```python
# Mdia de idade de cliente por bandeira

agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].mean().sort_values(ascending=False).reset_index(), 2)
agg_idade_bandeira
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='blue');
plt.ylabel('Mdia de idade')
plt.title('Idade Mdia por Bandeira', loc='center', fontsize='18', fontweight='bold', color='black')
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='blue')
plt.ylabel('Renda mdia por canal de venda')
plt.title('Renda mdia por canal de venda', fontsize='18', fontweight='bold', color='black')
```

```python
venda_por_data = vendas_correto.groupby('Data')['idcompra'].nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,8))
plt.xlabel('Data da Venda')
plt.ylabel('Qtd de Venda')
plt.title('Quantidade de Vendas por Ms')
x = venda_por_data['Data']
y = venda_por_data['idcompra']
plt.plot(x,y, color = 'blue', linewidth = '2')
plt.show()
```
